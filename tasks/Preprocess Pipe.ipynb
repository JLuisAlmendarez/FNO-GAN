```python
output_dir = '/media/joseluis/HDD/Proyecto CFD/datasets/Snapshots'

# --- CARGAR ---
data = np.load(
    '/media/joseluis/HDD/Proyecto CFD/datasets/Snapshots/snapshots_64x64.npz',
    mmap_mode='r'
)

vort = data['vorticity']
stream = data['stream']
vort_t = data['vort_t']
residual = data['residual']

# --- PARÁMETROS DE MUESTREO ---
n_samples = 100

np.random.seed(42)
indices = np.random.choice(len(vort), size=n_samples, replace=False)

# --- EXTRAER SUBSET ---
vort_sample = vort[indices]
stream_sample = stream[indices]
vort_t_sample = vort_t[indices]
residual_sample = residual[indices]

print(f'Subset shape vort:   {vort_sample.shape}')
print(f'Subset shape stream: {stream_sample.shape}')
print(f'Subset shape vort_t:   {vort_t_sample.shape}')
print(f'Subset shape residual: {residual_sample.shape}')


del data, vort, stream, vort_t, residual

np.savez_compressed(
    f'{output_dir}/snapshots_samples_64x64.npz',
    vorticity=vort_sample,
    stream=stream_sample,
    vort_t=vort_t_sample,
    residual=residual_sample
)

del vort_sample, stream_sample, vort_t_sample, residual_sample
```
